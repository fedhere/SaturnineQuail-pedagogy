# Pearson's Correlation. A measure of linear correlation.
![Pearson's Correlation Plot](https://upload.wikimedia.org/wikipedia/commons/thumb/d/d4/Correlation_examples2.svg/2560px-Correlation_examples2.svg.png)

\begin{align*}
    r_{xy} &= \frac{1}{n - 1} \sum^{N}_{i=1} \left( \frac{x_i - \bar{x}}{s_x} \right) \left( \frac{y_i - \bar{y}}{s_y} \right) \\
    \bar{x} &\equiv \textrm{mean value of x} \\
    \bar{y} &\equiv \textrm{mean value of y} \\
    n &\equiv \textrm{number of datapoints} \\
    s_x &\equiv \sqrt{\frac{1}{n - 1} \sum^{N}_{i=1}(x_i - \bar{x})^2}
\end{align*}



- "Correlation" vs "Anticorrelation" (Positively or negative correlation)
    - The Pearson correlation coefficient $r_{xy}$ is always between -1 and 1.
    - $r_{xy} > 0$ implies "correlation" ("positive correlation")
        - $r_{xy} = 1$ IF AND ONLY IF $y = |a|x$. "y and x are maximally correlated."
    - $r_{xy} < 0$ implies "anticorrelation" ("negative correlation")
        - $r_{xy} = -1$ IF AND ONLY IF $y = -|a|x$. "y and x are maximally anticorrelated."
- The bottom row in the figure represents correlation but not *linear* correlation, therefore $r_{xy} = 0$ even though $x$ and $y$ are clearly related to one another.

# Probability

## Frequentist vs Bayesian interpretation of probability

- Frequentist: "The fraction of times something happens is the probability of it happening."

\begin{align*}
    P(E) &= \textrm{frequency of event, E} \\
    \textrm{(After 11 coin flips) }P(\textrm{coin = heads}) &= \frac{6}{11} = 0.55 \\
    \textrm{(After 101 coin flips) }P(\textrm{coin = heads}) &= \frac{51}{101} = 0.504 \\
\end{align*}

- Bayesian: "I have a _prior_ belief about whether the coin is fair (50/50 of getting heads/tail) or not."
    - The Bayesian statistician could get 10 heads in a row, but if they have a prior belief that the coin is fair, then they will still expect to get 50% heads and 50% tails in the long run.

## Probability Arithmetic

### Rules

\begin{align*}
    \textrm{Probability is always between 0 and 1:  }& 0 <= P(A) <= 1 \\
    \textrm{If } \bar{A} \textrm{ is the complement of } A \textrm{:  }& P(A) + P(\bar{A}) = 1 \\
\end{align*}

**Disjoint events have independent probabilities.**

Two disjoint events: the intersection ($\cap$) of probabilities of event A and B is 0.

\begin{align*}
    \textrm{If }P(A) \cap P(B) = 0 \textrm{ then:}\\
    P(A \cup B) &= P(A) + P(B) \tag{1}\\
    P(A \cap B) &= P(A) * P(B) \tag{2}\\
    P(A | B) &= P(A) \tag{3}\\
\end{align*}

1. The probability of A or B ($A \cup B$) happening is the sum of the probabilities of A and B.
2. The probability of A and B ($A \cap B$) happening at once is the product of the probabilities of A and B.
3. The probability of A happening, given that B has happened, is simply the probability of A happening because A and B are disjoint (independent).


**Related events have dependent probabilities.**

Two related events: the intersection ($\cap$) of probabilities of event A and B is greater than 0.

\begin{align*}
    \textrm{If }P(A) \cap P(B) > 0 \textrm{ then:}\\
    P(A | B) &= \frac{P(A \cap B)}{P(B)} \tag{4}\\
    P(A | B) &< P(A) \tag{5}\\
    P(A \cap B) &= P(A) P(B|A) \tag{6}\\
\end{align*}

4. The probability of A happening, given that B has happened, is the probability of A and B happening ($A \cap B$) divided by the probability of B happening by itself.
5. The probability of A happening, given that B has happened, is always less than the probability of A happening by itself.
6. The probability of A and B happening ($A \cap B$) is the probability of A happening by itself times the probability of B happening given A has happened.

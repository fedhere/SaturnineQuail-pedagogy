# Null Hypothesis Rejection Testing

Objectives
* Calculate the Z-statistic of the sample of trip times.
* Calculate the KS-Test to determine whether two samples are from the same dsitribution.

```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
trip_duration = pd.read_csv(
    "https://raw.githubusercontent.com/fedhere/FDSfE_FBianco/refs/heads/main/data/triptimes.csv",
    header=None,
    names=["min"],
    )
trip_duration
```

# 1. Choose the Null Hypothesis

NH: Trip durations are distributed according to a Gaussian distribution with a mean of 34 minutes and standard deviation of 2.4 minutes.

```python
mu0 = 34
s0 = 2.4
N = trip_duration.size

mu0, s0, N
```

# 2. Identify all alternative outcomes

AH: The trip durations are NOT distributed according to a Gaussian distribution with a mean of 34 minutes and standard deviation of 2.4 minutes.

# 3. Set a confidence threshold

Choose a p-value of 0.05.

```python
p = 0.05
```

# 4. Find a measurable quantity such that the Null hypothesis has a known distribution.

We will use the Z-statistic.

# 5. Calculate the quantity.

Calculate the Z-statistic.

```python
Xbar = trip_duration["min"].mean()
Xbar
```

```python
Z = (Xbar - mu0) / (s0 / np.sqrt(N))
Z
```

# 6. Test data against alternative outcomes.

We chose a p-value of 0.05, making alpha = 95%. Calculate $p(NH|D)$

```python
from matplotlib import pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.hist(trip_duration["min"], density=True, label="Histogram of trip durations")

support = np.linspace(0, 75, num=1000)
pdf = stats.norm.pdf(support, loc=mu0, scale=s0)

ax.plot(support, pdf, label="Population distribution")

ax.legend()
fig.show()
```
